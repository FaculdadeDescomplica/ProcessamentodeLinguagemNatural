```python
documento = documento.content
documento = documento.replace("(? escutar)", "")
documento = documento.replace("\n", " ")
```

# Usando NLTK para segmentar os documentos

```python
import nltk
import string
from nltk.tokenize import word_tokenize, sent_tokenize
```

```python
nltk.download("punkt")
nltk.download("stopwords")
stopwords = nltk.corpus.stopwords.words('portuguese')
punctuation = string.punctuation
```

```python
tokens = word_tokenize(documento)
```

```python
print(tokens)
```

# Usando NLTK para obter a "raíz" das palavras

```python
from nltk.stem.snowball import SnowballStemmer
```

```python
stemmer = SnowballStemmer("portuguese")
```

```python
raizes = []

for token in tokens:
    if not token in punctuation and not token in stopwords:
        raizes.append(stemmer.stem(token))
```

```python
print(raizes)
```

```python
print(len(raizes))
print(len(set(raizes)))
```

# Contando as raízes mais frequentes

```python
frequencia_das_raizes = {
    raiz: raizes.count(raiz) for raiz in set(raizes)
    }
```

```python
print(sorted(frequencia_das_raizes.items(), key=lambda x: x[1], reverse=True))
```

```python
del frequencia_das_raizes["==="]
del frequencia_das_raizes["=="]
del frequencia_das_raizes["''"]
del frequencia_das_raizes["``"]
```

# Rankeando as frases do documento

```python
frases = sent_tokenize(documento)
```

```python
print(frases)
```

```python
ranking_frases = []

for frase in frases:
    pontos = 0
    tokens = word_tokenize(frase)
    raizes = [stemmer.stem(token) for token in tokens]
    for raiz in raizes:
        pontos += frequencia_das_raizes.get(raiz, 0)
    ranking_frases.append(pontos / len(tokens))
```

```python
print(ranking_frases)
```

# Retornando as frases com pontuação acima da média

```python
media_de_pontos = sum(ranking_frases) / len(ranking_frases)
media_de_pontos
```

```python
resumo = []
for i, frase in enumerate(frases):
    pontos = ranking_frases[i]
    if pontos > (1.2 * media_de_pontos):
        resumo.append(frase)
    if len(resumo) == 5:
        resumo = " ".join(resumo)
        break
```
