# Abrindo a legenda e entendendo sua estrutura

```python
with open("breaking-bad-s01e01.txt") as f:
    legenda = f.read()
```

```python
print(legenda)
```

# Transformando a legenda em um só texto

```python
falas = legenda.split("\n\n")
```

```python
falas
```

```python
falas[0]
```

```python
falas[0].split("\n", 2)
```

```python
falas[0].split("\n", 2)[2]
```

```python
texto = ""

for fala in falas:
    texto = texto + " " + fala.split("\n", 2)[2]
```

```python
texto
```

# Limpando a legenda

```python
texto = texto.replace("\n", " ")
texto = texto.replace("<b>", "")
texto = texto.replace("</b>", "")
texto = texto.replace("<i>", "")
texto = texto.replace("</i>", "")
```

```python
texto
```

# Segmentando frases

```python
frases = texto.split(". ")
```

```python
frases
```

```python
frases = [frase + "." for frase in frases]
```

```python
frases
```

# Segmentando tokens

```python
# List comprehension: laço de repetição (loop) dentro de uma lista para gerar outra lista
frases = [frase\
          .replace(".", " . ")\
          .replace(",", " , ")\
          .replace("!", " ! ")\
          .replace("?", " ? ")\
          .replace(":", " : ")\
          .replace("  ", " ")\
          .strip() for frase in frases]
```

```python
frases
```

```python
frases_tokenizadas = [frase.split(" ") for frase in frases]
```

```python
frases_tokenizadas
```

```python
frases_tokenizadas[0]
```

# Gerando lista de palavras-chave usando YAKE


```python
! pip3 install yake
```

```python
import yake
```

```python
texto
```

```python
kw_extractor = yake.KeywordExtractor(lan="pt", n=1)
keywords = kw_extractor.extract_keywords(texto)
```

```python
keywords
```

```python
sorted(keywords, key=lambda x: x[1], reverse=True)
```
