# Vídeo 3: Tokenizando manualmente

```python
frases_nltk
```

```python
frases_tokenizadas = []

for frase in frases_nltk:
    frase = frase.replace(".", " . ")
    frase = frase.replace(",", " , ")
    frase = frase.replace(":", " : ")
    frase = frase.replace("(", " ( ")
    frase = frase.replace(")", " ) ")
    frase = frase.replace("-", " - ")
    frase = frase.replace("«", " « ")
    frase = frase.replace("»", " » ")
    frase = frase.replace("  ", " ")
    frase = frase.strip()
    frases_tokenizadas.append(frase.split(" "))
```

```python
frases_tokenizadas
```

```python
frases_tokenizadas[8]
```

# Tokenizando usando NLTK

```python
frases_nltk
```

```python
frases_tokenizadas_nltk = []

for frase in frases_nltk:
    frase = nltk.tokenize.word_tokenize(frase)
    frases_tokenizadas_nltk.append(frase)
```

```python
frases_tokenizadas_nltk
```

```python
frases_tokenizadas_nltk[8]
```

# Vídeo 4: Removendo stopwords manualmente

```python
! wget -O stopwords.txt https://raw.githubusercontent.com/stopwords-iso/stopwords-pt/master/stopwords-pt.txt
```

```python
with open("stopwords.txt") as f:
    stopwords = f.read()
```

```python
print(stopwords)
```

```python
stopwords = stopwords.split("\n")
```

```python
stopwords
```

```python
frases_tokenizadas_nltk
```

```python
frases_sem_stopwords = []

for frase in frases_tokenizadas_nltk:
    frase_sem_stopwords = []
    for token in frase:
        if not token.lower() in stopwords:
            frase_sem_stopwords.append(token)
    frases_sem_stopwords.append(frase_sem_stopwords)
```

```python
frases_sem_stopwords
```

# Removendo stopwords usando NLTK

```python
import string
string.punctuation
```

```python
nltk.download("stopwords")
```

```python
nltk.corpus.stopwords.words('portuguese')
```

```python
frases_sem_stopwords_nltk = []

for frase in frases_tokenizadas_nltk:
    frase_sem_stopwords = []
    for token in frase:
        if not token.lower() in nltk.corpus.stopwords.words('portuguese') and not token in string.punctuation:
            frase_sem_stopwords.append(token)
    frases_sem_stopwords_nltk.append(frase_sem_stopwords)
```
