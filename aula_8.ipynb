# Utilizando NLTK

```python
! pip3 install nltk
! pip3 install svgling
```

```python
import nltk
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('maxent_ne_chunker')
nltk.download('words')
```

```python
from nltk.tokenize import word_tokenize
from nltk.tag import pos_tag
from nltk.chunk import ne_chunk
```

```python
tokens = word_tokenize(pagina.content)
```

```python
print(tokens)
```

```python
tags = pos_tag(tokens)
```

```python
print(tags)
```

```python
entities = ne_chunk(tags)
```

```python
entities
```
