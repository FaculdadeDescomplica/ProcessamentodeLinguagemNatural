```python
from datasets import load_dataset
from transformers import pipeline
```

```python
dataset = load_dataset("Censius-AI/ECommerce-Women-Clothing-Reviews")
```

```python
dataset
```

```python
avaliacoes = dataset['train']['Review Text']
```

```python
classificador = pipeline(
    model="lxyuan/distilbert-base-multilingual-cased-sentiments-student",
    return_all_scores=True,
)
```

```python
classificador("Achei o filme excelente")
```

```python
classificador("Achei o filme horrível!")
```

```python
classificador("Achei o filme ok")
```

```python
classificador("Não sei dizer se gostei do filme")
```

```python
classificador(["Não sei dizer se gostei do filme", "Adorei o filme"])
```

```python
print([x for x in avaliacoes if not x])
```

```python
avaliacoes = [x if x else "" for x in avaliacoes]
```

```python
print([x for x in avaliacoes if not x])
```

```python
# Para todas as 23.486 avaliações, a classificação demoraria 1 hora
classificacoes = classificador(avaliacoes[:100])
```

```python
for i, avaliacao in enumerate(avaliacoes[:100]):
    if avaliacao:
        print((estrelas[i], classificacoes[i]))
```
