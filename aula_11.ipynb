```python
from datasets import load_dataset
from textblob import TextBlob
```

```python
dataset = load_dataset("Censius-AI/ECommerce-Women-Clothing-Reviews")
```

```python
dataset
```

```python
avaliacoes = dataset['train']['Review Text']
```

```python
avaliacoes[0]
```

```python
estrelas = dataset['train']['Rating']
```

```python
estrelas[0]
```

```python
len(avaliacoes)
```

```python
len(estrelas)
```

```python
TextBlob(avaliacoes[0]).sentiment
```

```python
for i, avaliacao in enumerate(avaliacoes):
    if avaliacao:
        polaridade = TextBlob(avaliacao).sentiment.polarity
        print(estrelas[i], polaridade)
```

```python
sentimentos = {"positivo": [5, 4], "neutro": [3], "negativo": [1, 2]}
analises_estranhas = []
for i, avaliacao in enumerate(avaliacoes):
    if avaliacao:
        polaridade = TextBlob(avaliacao).sentiment.polarity
        if polaridade > 0.2:
            sentimento = "positivo"
        elif -0.2 <= polaridade <= 0.2:
            sentimento = "neutro"
        elif polaridade < -0.2:
            sentimento = "negativo"
        avaliacao_estrelas = estrelas[i]
        if avaliacao_estrelas not in sentimentos[sentimento]:
            analises_estranhas.append((estrelas[i], polaridade))
```

```python
print(analises_estranhas)
```

```python
len(analises_estranhas)
```

```python
len(analises_estranhas) / len(avaliacoes)
```
